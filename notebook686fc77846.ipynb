```python
df = pd.read_csv('/kaggle/input/myntra-fashion-dataset/Myntra Fasion Clothing.csv')
df.head()
```

```python
from transformers import AutoTokenizer, AutoModel
model_name = 'microsoft/codebert-base'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
from transformers import BertTokenizer, ErnieModel

tokenizer = BertTokenizer.from_pretrained("nghuyong/ernie-3.0-nano-zh")
model = ErnieModel.from_pretrained("nghuyong/ernie-3.0-nano-zh")
```

```python
sentences = ["black t shirt for women"] + list(df['Description'][:50].values)

tokenized = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')
outputs = model(**tokenized)
sentence_embeddings = outputs.last_hidden_state.mean(dim=1)
```

```python
!pip install sentence_transformers
from sentence_transformers import SentenceTransformer
sentences = ["black t shirt for women"] + list(df['Description'][:1000].values)

model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
sentence_embeddings = model.encode(sentences)
```

```python
sentence_embeddings
```

```python
import math

def dot_product(vec1, vec2):
    return sum(x * y for x, y in zip(vec1, vec2))

def magnitude(vec):
    return math.sqrt(sum(x * x for x in vec))

def cosine_similarity(vec1, vec2):
    dot_prod = dot_product(vec1, vec2)
    mag1 = magnitude(vec1)
    mag2 = magnitude(vec2)
    
    if mag1 != 0 and mag2 != 0:
        return dot_prod / (mag1 * mag2)
    else:
        return 0.0  # Handle zero vectors

similarities = []
for i in range(1000):
    similarity = cosine_similarity(sentence_embeddings[0], sentence_embeddings[i + 1])
    similarities.append([df['Description'].iloc[i],similarity.item()])
    #print("Cosine Similarity:", similarity)
sorted(similarities,key = (lambda x : -x[1]))[:10]
```
